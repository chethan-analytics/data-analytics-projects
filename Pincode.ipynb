```python
import pandas as pd
import requests
import time

# ------------------- User Config -------------------
input_file = r"C:\Users\chethan.m\Downloads\input.xlsx"  # your input Excel path
output_file = r"C:\Users\chethan.m\Downloads\output west.xlsx"
# ---------------------------------------------------

# Read input Excel
df = pd.read_excel(input_file, dtype=str)

headers = {"User-Agent": "Mozilla/5.0"}

# New list to store rows in final Excel
output_rows = []

for idx, row in df.iterrows():
    pincode = row["Pincode"].strip()
    print(f"\nFetching areas for Pincode {idx+1}/{len(df)}: {pincode}")

    attempts = 0
    max_attempts = 3
    success = False

    while attempts < max_attempts and not success:
        try:
            url = f"https://api.postalpincode.in/pincode/{pincode}"
            response = requests.get(url, headers=headers, timeout=20).json()
            data = response[0]

            if data["Status"] == "Success" and data["PostOffice"]:
                areas = [po["Name"] for po in data["PostOffice"]]

                # Add each area as a separate row
                for i, area in enumerate(areas, start=1):
                    output_rows.append({"Pincode": pincode, "Area": area})
                    print(f"{i}. {area} ({pincode})")

                success = True
            else:
                output_rows.append({"Pincode": pincode, "Area": "No data found"})
                print("No data found for this pincode.")
                success = True

        except Exception as e:
            attempts += 1
            print(f"Attempt {attempts} failed: {e}")
            time.sleep(2)

    time.sleep(1)

# Create final DataFrame and save to Excel
output_df = pd.DataFrame(output_rows)
output_df.to_excel(output_file, index=False)
print(f"\nDone! Check {output_file}")
```
